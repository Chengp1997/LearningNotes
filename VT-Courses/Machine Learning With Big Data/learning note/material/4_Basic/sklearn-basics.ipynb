# Scikit-Learn Basics


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 


import sklearn
from sklearn import linear_model
```

## Create and load data

```python
n = 200
A = 1.5
B = 13
x = np.random.random(n) * 10  # generate a uniform distribution from (0, 10] 正态分布
x = x.reshape(-1, 1)   # sklearn expects a variable to be column-oriented so we make num_instances by num_features  变成只有一列的
# print(x)
ytrue = A * x + B  # True values
yrand = ytrue + np.random.normal(0, 3, size=n).reshape(-1, 1)
```

```python
plt.scatter(x, ytrue)
plt.scatter(x, yrand)
```

```python
sklearn.metrics.mean_squared_error(ytrue, yrand)  #均方误差
```

## A Simple Model

```python
reg = linear_model.LinearRegression()
print(reg)
```

```python
reg.fit(x, yrand)
```

```python
reg.coef_, reg.intercept_ #回归系数，截距
```

```python
plt.scatter(x, yrand)

ypred = reg.predict(x)  #进行一个模拟，模拟一下大致的曲线
plt.plot(x, ypred, color='r');
```

```python
sklearn.metrics.r2_score(ytrue, ypred) # evaluation, 使用r^2评估一下模型。
```
